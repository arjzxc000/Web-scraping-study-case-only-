```python
# click on datepicker
WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH, "/html/body/app-root/app-body/div/app-available-assets/app-search-basic/div/div/div/div/div[1]/div/div[2]/div[2]/div/input"))).click()
# click on year selector
WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="owl-dt-picker-2"]/div[2]/owl-date-time-calendar/div[1]/div/button/span'))).click()
# click on year
WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="owl-dt-picker-2"]/div[2]/owl-date-time-calendar/div[2]/owl-date-time-multi-year-view/table/tbody/tr[2]/td[3]/span'))).click()
# click on month
WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="owl-dt-picker-2"]/div[2]/owl-date-time-calendar/div[2]/owl-date-time-year-view/table/tbody/tr[4]/td[2]/span'))).click()
# click on date
WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="owl-dt-picker-2"]/div[2]/owl-date-time-calendar/div[2]/owl-date-time-month-view/table/tbody/tr[3]/td[2]/span'))).click()
# click set
WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="owl-dt-picker-2"]/div[2]/div[2]/button[2]/span'))).click()
# click on date
WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="owl-dt-picker-3"]/div[2]/owl-date-time-calendar/div[2]/owl-date-time-month-view/table/tbody/tr[3]/td[3]/span'))).click()
# click set
WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="owl-dt-picker-3"]/div[2]/div[2]/button[2]/span'))).click()
# click search
search_button = driver.find_element_by_xpath('/html/body/app-root/app-body/div/app-available-assets/app-search-basic/div/div/div/div/div[2]')
search_button.click()
```

```python
# search_button = driver.find_element_by_xpath('/html/body/app-root/app-body/div/app-available-assets/app-search-basic/div/div/div/div/div[2]')
# search_button.click()
while True:
    try:
        time.sleep(1.5)
        html = driver.page_source
        s = BeautifulSoup(html,'html.parser')
        # == result and owners status == #
        results = s.find_all('a',{'class':'card-detail'})
        owners = s.find_all('span',{'class':'username'})
        owners_status = s.find_all('span',{'class':'badge'})
        img_block = s.find_all('div',{'class':'car-thumb'})
        # == data == #
        for index,v in enumerate(results):
            name = results[index].find_all('h6',{'class':'card-title'})[0].text.strip()[:-5]
            year = results[index].find_all('h6',{'class':'card-title'})[0].text.strip()[-4:]
            location = results[index].find_all('span',{'class':'location'})[0].text.strip()
            type = results[index].find_all('li')[0].text.strip()
            driving_system = results[index].find_all('li')[1].text.strip()
            seat = results[index].find_all('li')[2].text.strip()
            price_per_day =  results[index].find_all('span',{'class':'card-price'})[0].text.strip()[:-3]
            owner = owners[index].text.strip()
            img = img_block[index].find('img')['src']
            # owner_status = owners_status[index].text.strip()
            car_name.append(name)
            car_year.append(year)
            car_location.append(location)
            car_type.append(type)
            car_driving_system.append(driving_system)
            car_seat.append(seat)
            car_price_per_day.append(price_per_day)
            car_owner.append(owner)
            car_image.append(img)
        # ==== click next button ==== #
        next_button = driver.find_element_by_xpath('/html/body/app-root/app-body/div/app-available-assets/div[1]/div/div/div[2]/div[2]/nav/div/div[3]/a')
        driver.execute_script("arguments[0].click();",next_button)
    except:
        break
```

```python
data = dict(
    car_name = car_name,
    car_year = car_year,
    car_location = car_location,
    car_type =car_type,
    car_driving_system = car_driving_system,
    car_seat = car_seat,
    car_price_per_day = car_price_per_day,
    car_owner = car_owner,
    car_image = car_image
)
df = pd.DataFrame(data)
```
